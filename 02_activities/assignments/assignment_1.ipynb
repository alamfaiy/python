### Part 1: Building the base Anagram Checker

Given two valid strings, check to see if they are anagrams of each other. If it is, return `True`, else `False`. For this part, we can assume that uppercase letters are the same as if it was a lowercase character.

Examples of anagrams:
* Silent and Listen
* Night and Think

Example outputs:
```python
anagram_checker("Silent", "listen") # True
anagram_checker("Silent", "Night") # False
anagram_checker("night", "Thing") # True
```

````python
# For testing purposes, we will write our code in the function
def anagram_checker(word_a, word_b):
    # First I wanted to make sure the word lengths actually matched
    if len(word_a) == len(word_b): 
      #print("Anagrams are possible") # I was testing if the initial if statement worked

      # If word lengths match, remove the case sensitivity (make it all lowercase)
      word_a = word_a.lower()
      word_b = word_b.lower()

      # Then, splice the words into seperate letters to compare later
      # First, create the empty sets to put my letters in
      a_letters = []
      b_letters = []

      # For every single letter in the word (doesn't matter which, start with word_a)
      for letter in range(len(word_a)):

        # Extract out the letters
        letter_a = word_a[letter]
        letter_b = word_b[letter]

        # Append the letters into the list
        a_letters.append(letter_a)
        b_letters.append(letter_b)
    
      # Verify that the sorted letters are lowercase  
      #print(sorted(a_letters))
      #print(sorted(b_letters))

      # Sort the letters in each list first
      # If the lists are identical after sorting, they are anagrams!
      # Note: if my lists contained capital letters it would sort differently
      if sorted(a_letters) == sorted(b_letters):
        print('True')
        print('These two words make an anagram!')
      # If the lists are not identical after sorting, they are not anagrams
      else:
        print('False')
        print('These two words do not make an anagram')

    # If word lengths don't match, say so
    else: 
      print("Different word lengths can't be an anagram")


# Run your code to check using the words below:
anagram_checker("Silent", "listen")
````

````python
anagram_checker("Silent", "Night")
````

````python
anagram_checker("night", "Thing")
````
